```python
import pandas as pd
import numpy as np
import seaborn as sea
import matplotlib.pyplot as plt
%matplotlib inline
import math
pokemon_dataset=pd.read_csv("Pokemon_database.csv")
pokemon_dataset.head(9)
```

```python
print("No. of pokemon in original dataset:"+str(len(pokemon_dataset.index)))
```

```python
pokemon_dataset.drop(['Pokemon Id','Pokemon Name','Pokedex Number','Alternate Form Name','Original Pokemon ID'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Secondary Type','Primary Ability','Primary Ability Description','Secondary Ability'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Hidden Ability','Hidden Ability Description','Special Event Ability','Special Event Ability Description','Secondary Ability Description'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Base Happiness','Game(s) of Origin','Region of Origin','Health Stat','Attack Stat','Defense Stat'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Special Attack EV','Special Defense EV','Speed EV','Catch Rate'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Legendary Type','Male Ratio','Female Ratio','Special Attack Stat','Special Defense Stat'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Experience Growth','Experience Growth Total','Primary Egg Group','Secondary Egg Group'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Egg Cycle Count','Pre-Evolution Pokemon Id','Evolution Details'],axis=1,inplace=True)
```

```python
pokemon_dataset.drop(['Primary Type','Base Stat Total','Health EV','Attack EV','Defense EV'],axis=1,inplace=True)
```

```python
from pandas import DataFrame
df=DataFrame(pokemon_dataset,columns=['Pokemon Height','Pokemon Weight','Speed Stat','EV Yield Total','Experience Yield'])
df.head(8)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
df,pokemon_dataset['Classification'],random_state=0)
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
#for testing the model
X_new = np.array([[1.2,19.1,81,2,143]])
```

```python
prediction = knn.predict(X_new)
print("Prediction:{}".format(prediction))
```

```python
print("Test set score:{}".format(knn.score(X_test,y_test)*100))
```
